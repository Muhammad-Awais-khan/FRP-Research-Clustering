```python
import pandas as pd
import os
```

```python
data_files_names = os.listdir('data/')
print(data_files_names)
```

```python
data = pd.DataFrame()
for file_name in data_files_names:
    files_path = os.path.join('data/', file_name)
    temp_data = pd.read_csv(files_path)
    data = pd.concat([data, temp_data], ignore_index=True)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
abstract_embeddings = model.encode(abstract, show_progress_bar=True)
```
